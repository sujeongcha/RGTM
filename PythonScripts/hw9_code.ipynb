```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
rs_1 = np.loadtxt("radioactive_sample_1.txt")
ma_df = pd.DataFrame({"raw_data":rs_1})
ma_df["cam"] = ma_df["raw_data"].expanding().mean()
```

```python
plt.rcParams["figure.figsize"] = (10,5)
plt.plot(ma_df["cam"])
plt.xlabel("Sample Size")
plt.ylabel("Moving Average")
plt.yticks(np.arange(2.6, 3.7, 0.1))
plt.axhline(y=3.5, color="r", linestyle="--")
```
