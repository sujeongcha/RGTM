```python
import pandas as pd
from sodapy import Socrata

client = Socrata("data.cityofnewyork.us", 
                 "phCy46lgZSogk7QgnXpSoMkSl",
                 username="sujeongcha90@gmail.com",
                 password="Housing311")
```

```python
# Loading data for Manhattan (separate load by year due to large data volume)
Manhattan_2015 = client.get("erm2-nwe9", 
                            where="borough='MANHATTAN' AND created_date between '2015-01-01T00:00:00.000' and '2015-12-31T23:59:59.000'", limit=500000)
Manhattan_2016 = client.get("erm2-nwe9", 
                            where="borough='MANHATTAN' AND created_date between '2016-01-01T00:00:00.000' and '2016-12-31T23:59:59.000'", limit=500000)
Manhattan_2017 = client.get("erm2-nwe9", 
                            where="borough='MANHATTAN' AND created_date between '2017-01-01T00:00:00.000' and '2017-12-31T23:59:59.000'", limit=500000)
Manhattan_2018 = client.get("erm2-nwe9", 
                            where="borough='MANHATTAN' AND created_date between '2018-01-01T00:00:00.000' and '2018-12-31T23:59:59.000'", limit=500000)
Manhattan_2019 = client.get("erm2-nwe9", 
                            where="borough='MANHATTAN' AND created_date between '2019-01-01T00:00:00.000' and '2019-12-31T23:59:59.000'", limit=500000)
```

```python
# Appending the dataframes row-wise into one dataframe

comp2015 = pd.DataFrame(Manhattan_2015)
comp2016 = pd.DataFrame(Manhattan_2016)
comp2017 = pd.DataFrame(Manhattan_2017)
comp2018 = pd.DataFrame(Manhattan_2018)
comp2019 = pd.DataFrame(Manhattan_2019)

complaints_man = comp2015.append([comp2016, comp2017, comp2018, comp2019], sort=False)
pd.set_option('display.max_columns', 500)
display(complaints_man.head())
```

```python
# Restricting Complaint types to
# Rodent/Hazardous Materials/Sidewalk Condition/WATER LEAK/Noise - Residential/Noise - Commercial

condition = ['Rodent', 'Hazardous Materials', 'Sidewalk Condition', 'WATER LEAK', 'Noise - Residential', 'Noise - Commercial']

def drop_complaints(data):
    result = data[data['complaint_type'].isin(condition)]
    return result

six_comp_man = drop_complaints(complaints_man)
display(six_comp_man.head())
```

```python
# Cleaning up columns

comp_cleaned = six_comp_man[['complaint_type', 'latitude', 'longitude']]
comp_cleaned['longlat'] = list(zip(comp_cleaned.longitude, comp_cleaned.latitude))
display(comp_cleaned.head())
```
