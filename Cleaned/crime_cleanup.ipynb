```python
# Loading data for Manhattan (separate load by year due to large data volume)
Manhattan_2015 = client.get("qgea-i56i", 
                            where="boro_nm='MANHATTAN' AND cmplnt_fr_dt between '2015-01-01T00:00:00.000' and '2015-12-31T23:59:59.000'", limit=500000)
Manhattan_2016 = client.get("qgea-i56i", 
                            where="boro_nm='MANHATTAN' AND cmplnt_fr_dt between '2016-01-01T00:00:00.000' and '2016-12-31T23:59:59.000'", limit=500000)
Manhattan_2017 = client.get("qgea-i56i", 
                            where="boro_nm='MANHATTAN' AND cmplnt_fr_dt between '2017-01-01T00:00:00.000' and '2017-12-31T23:59:59.000'", limit=500000)
Manhattan_2018 = client.get("qgea-i56i", 
                            where="boro_nm='MANHATTAN' AND cmplnt_fr_dt between '2018-01-01T00:00:00.000' and '2018-12-31T23:59:59.000'", limit=500000)
Manhattan_2019 = client.get("qgea-i56i", 
                            where="boro_nm='MANHATTAN' AND cmplnt_fr_dt between '2019-01-01T00:00:00.000' and '2019-12-31T23:59:59.000'", limit=500000)
```

```python
# Appending the dataframes row-wise into one dataframe

crime2015 = pd.DataFrame(Manhattan_2015)
crime2016 = pd.DataFrame(Manhattan_2016)
crime2017 = pd.DataFrame(Manhattan_2017)
crime2018 = pd.DataFrame(Manhattan_2018)
crime2019 = pd.DataFrame(Manhattan_2019)

crime_man = crime2015.append([crime2016, crime2017, crime2018, crime2019], sort=False)
pd.set_option('display.max_columns', 500)
display(crime_man.head())
```

```python
# Restricting complaint types to FELONY                                 

condition = ['FELONY']

def drop_complaints(data):
    result = data[data['law_cat_cd'].isin(condition)]
    return result

felony_man = drop_complaints(crime_man)
display(felony_man.head())
```

```python
# Cleaning up columns

crime_cleaned = felony_man[['ofns_desc', 'latitude', 'longitude']]
crime_cleaned['GeoCode'] = list(zip(crime_cleaned.longitude, crime_cleaned.latitude))
display(crime_cleaned.head())
```
