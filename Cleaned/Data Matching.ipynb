```python
import pandas as pd

nyc_open = pd.read_csv('~/Desktop/workspace/FinalProject/RGTM-master/Cleaned/merged_manhattan_cleanup.csv', header = 0)
zillow = pd.read_csv('~/Desktop/workspace/FinalProject/merged_zillow_cleanup.csv', header = 0)
```

```python
# Fill in missing values for AptNum
nyc_open['AptNum'] = nyc_open['AptNum'].fillna('House')
zillow['AptNum'] = zillow['AptNum'].fillna('House')

# Remove all the spaces
nyc_open['AptNum'] = nyc_open['AptNum'].str.strip()
zillow['AptNum'] = zillow['AptNum'].str.strip()
nyc_open['StreetNum'] = nyc_open['StreetNum'].str.strip()
zillow['StreetNum'] = zillow['StreetNum'].str.strip()
nyc_open['StreetName'] = nyc_open['StreetName'].str.strip()
zillow['StreetName'] = zillow['StreetName'].str.strip()

# Drop rows with no AptNum
indexNames = nyc_open[nyc_open['AptNum'].isna()].index
nyc_open = nyc_open.drop(indexNames)
print(nyc_open.shape)
indexNames = zillow[zillow['AptNum'].isna()].index
zillow = zillow.drop(indexNames)
print(zillow.shape)
```

```python
# Match apartments 
matched = pd.merge(nyc_open, zillow, on = ['StreetNum', 'StreetName', 'AptNum', 'ZipCode'])
matched
```

```python
matched.to_csv('~/Desktop/workspace/FinalProject/matched_data_cleanup.csv', index=False)
```
